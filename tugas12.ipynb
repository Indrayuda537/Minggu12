```python
# ============================================================
# SISTEM PENCOCOKAN OBJEK BERBASIS FITUR LOKAL
# ============================================================

import cv2
import numpy as np
import os
import time
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    classification_report,
    ConfusionMatrixDisplay,
    precision_recall_curve,
    average_precision_score
)

# ============================================================
# FOLDER SAAT INI
# ============================================================

CURRENT_FOLDER = "."

# ============================================================
# GENERATE DATASET
# ============================================================

def generate_dataset():

    files = [
        f for f in os.listdir(CURRENT_FOLDER)
        if f.lower().endswith((".jpg", ".png", ".jpeg"))
    ]

    for file in files:

        if any(x in file for x in [
            "_ref",
            "_rotasi",
            "_gelap",
            "_zoom",
            "_oklusi",
            "_blur"
        ]):
            continue

        name = os.path.splitext(file)[0]

        img = cv2.imread(file)

        if img is None:
            continue

        h, w = img.shape[:2]

        # ====================================================
        # REF
        # ====================================================

        cv2.imwrite(f"{name}_ref.jpg", img)

        # ====================================================
        # ROTASI
        # ====================================================

        M = cv2.getRotationMatrix2D(
            (w//2, h//2),
            30,
            1
        )

        rot = cv2.warpAffine(img, M, (w, h))

        cv2.imwrite(f"{name}_rotasi.jpg", rot)

        # ====================================================
        # GELAP
        # ====================================================

        dark = cv2.convertScaleAbs(
            img,
            alpha=0.6,
            beta=-40
        )

        cv2.imwrite(f"{name}_gelap.jpg", dark)

        # ====================================================
        # ZOOM
        # ====================================================

        zoom = cv2.resize(
            img,
            None,
            fx=1.5,
            fy=1.5
        )

        zh, zw = zoom.shape[:2]

        x = (zw - w)//2
        y = (zh - h)//2

        zoom = zoom[y:y+h, x:x+w]

        cv2.imwrite(f"{name}_zoom.jpg", zoom)

        # ====================================================
        # OKLUSI
        # ====================================================

        occ = img.copy()

        cv2.rectangle(
            occ,
            (w//3, h//3),
            (w//2, h//2),
            (0,0,0),
            -1
        )

        cv2.imwrite(f"{name}_oklusi.jpg", occ)

        # ====================================================
        # BLUR
        # ====================================================

        blur = cv2.GaussianBlur(
            img,
            (9,9),
            0
        )

        cv2.imwrite(f"{name}_blur.jpg", blur)

    print("\nDataset berhasil dibuat!")

# ============================================================
# LOAD DATASET
# ============================================================

def load_dataset():

    images = []
    labels = []
    paths = []

    files = [
        f for f in os.listdir(CURRENT_FOLDER)
        if any(x in f for x in [
            "_ref",
            "_rotasi",
            "_gelap",
            "_zoom",
            "_oklusi",
            "_blur"
        ])
    ]

    for file in files:

        img = cv2.imread(file, 0)

        if img is None:
            continue

        label = file.split("_")[0]

        images.append(img)
        labels.append(label)
        paths.append(file)

    return images, labels, paths

# ============================================================
# DETECTOR
# ============================================================

def get_detector(method):

    if method == "SIFT":
        return cv2.SIFT_create()

    elif method == "ORB":
        return cv2.ORB_create(nfeatures=1000)

    elif method == "SURF":

        try:
            return cv2.xfeatures2d.SURF_create()

        except:
            return None

# ============================================================
# FEATURE EXTRACTION
# ============================================================

def extract_features(images, method):

    detector = get_detector(method)

    if detector is None:

        print(f"\n{method} tidak tersedia")
        return None

    kp_list = []
    des_list = []

    total_time = 0

    print(f"\n===== FEATURE EXTRACTION {method} =====")

    for i, img in enumerate(images):

        start = time.time()

        kp, des = detector.detectAndCompute(
            img,
            None
        )

        end = time.time()

        elapsed = end - start

        total_time += elapsed

        if des is None:
            des = np.array([])

        kp_list.append(kp)
        des_list.append(des)

        print(f"\nGambar {i+1}")

        print("Jumlah Keypoints :", len(kp))

        if len(des) > 0:
            print("Descriptor Shape :", des.shape)

        print(f"Waktu : {elapsed:.4f} detik")

    print(f"\nTotal Waktu {method}: {total_time:.4f}")

    return kp_list, des_list

# ============================================================
# VISUALISASI KEYPOINT
# ============================================================

def visualize_keypoints(img, kp, title):

    result = cv2.drawKeypoints(
        img,
        kp,
        None,
        flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS
    )

    plt.figure(figsize=(8,6))

    plt.imshow(result, cmap='gray')

    plt.title(title)

    plt.axis("off")

    plt.show()

# ============================================================
# BF MATCHING
# ============================================================

def bf_matching(method, des1, des2):

    if len(des1) == 0 or len(des2) == 0:
        return []

    if method in ["SIFT", "SURF"]:

        bf = cv2.BFMatcher(cv2.NORM_L2)

    else:

        bf = cv2.BFMatcher(cv2.NORM_HAMMING)

    matches = bf.knnMatch(des1, des2, k=2)

    good = []

    for pair in matches:

        if len(pair) == 2:

            m, n = pair

            if m.distance < 0.75 * n.distance:
                good.append(m)

    return good

# ============================================================
# FLANN MATCHING
# ============================================================

def flann_matching(method, des1, des2):

    if len(des1) == 0 or len(des2) == 0:
        return []

    if method in ["SIFT", "SURF"]:

        index_params = dict(
            algorithm=1,
            trees=5
        )

        search_params = dict(checks=50)

    else:

        index_params = dict(
            algorithm=6,
            table_number=6,
            key_size=12,
            multi_probe_level=1
        )

        search_params = dict(checks=50)

        des1 = np.uint8(des1)
        des2 = np.uint8(des2)

    flann = cv2.FlannBasedMatcher(
        index_params,
        search_params
    )

    matches = flann.knnMatch(
        des1,
        des2,
        k=2
    )

    good = []

    for pair in matches:

        if len(pair) == 2:

            m, n = pair

            if m.distance < 0.75 * n.distance:
                good.append(m)

    return good

# ============================================================
# RANSAC
# ============================================================

def ransac_homography(kp1, kp2, matches):

    if len(matches) < 4:
        return None, None

    src_pts = np.float32([
        kp1[m.queryIdx].pt
        for m in matches
    ]).reshape(-1,1,2)

    dst_pts = np.float32([
        kp2[m.trainIdx].pt
        for m in matches
    ]).reshape(-1,1,2)

    H, mask = cv2.findHomography(
        src_pts,
        dst_pts,
        cv2.RANSAC,
        5.0
    )

    return H, mask

# ============================================================
# DRAW MATCHES
# ============================================================

def draw_matches(
    img1,
    kp1,
    img2,
    kp2,
    matches,
    mask=None
):

    matchesMask = (
        mask.ravel().tolist()
        if mask is not None
        else None
    )

    result = cv2.drawMatches(
        img1,
        kp1,
        img2,
        kp2,
        matches,
        None,
        matchesMask=matchesMask,
        flags=2
    )

    plt.figure(figsize=(15,8))

    plt.imshow(result)

    plt.axis("off")

    plt.show()

# ============================================================
# PRECISION RECALL
# ============================================================

def precision_recall_analysis(matches):

    if len(matches) == 0:
        return

    distances = np.array([
        m.distance for m in matches
    ])

    y_true = np.ones(len(distances))

    scores = -distances

    precision, recall, _ = precision_recall_curve(
        y_true,
        scores
    )

    ap = average_precision_score(
        y_true,
        scores
    )

    plt.figure(figsize=(7,5))

    plt.plot(recall, precision)

    plt.xlabel("Recall")

    plt.ylabel("Precision")

    plt.title(f"Precision Recall Curve\nAP={ap:.2f}")

    plt.grid()

    plt.show()

# ============================================================
# BOVW
# ============================================================

def build_bovw(descriptors_list, k=20):

    all_descriptors = []

    for des in descriptors_list:

        if len(des) > 0:

            for d in des:

                all_descriptors.append(d)

    all_descriptors = np.array(all_descriptors)

    kmeans = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10
    )

    kmeans.fit(all_descriptors)

    return kmeans

# ============================================================
# HISTOGRAM VISUAL WORD
# ============================================================

def extract_histogram(kmeans, descriptors, k):

    hist = np.zeros(k)

    if len(descriptors) == 0:
        return hist

    clusters = kmeans.predict(descriptors)

    for c in clusters:
        hist[c] += 1

    return hist

# ============================================================
# CLASSIFICATION
# ============================================================

def classification(X, y):

    # ========================================================
    # TRAIN TEST SPLIT
    # ========================================================

    X_train, X_test, y_train, y_test = train_test_split(
        X,
        y,
        test_size=0.3,
        random_state=42,
        stratify=y
    )

    # ========================================================
    # SCALING
    # ========================================================

    scaler = StandardScaler()

    X_train = scaler.fit_transform(X_train)

    X_test = scaler.transform(X_test)

    # ========================================================
    # LABEL SEMUA KELAS
    # ========================================================

    all_labels = sorted(list(set(y)))

    # ========================================================
    # SVM
    # ========================================================

    svm = SVC()

    svm.fit(X_train, y_train)

    pred_svm = svm.predict(X_test)

    svm_acc = accuracy_score(
        y_test,
        pred_svm
    )

    print("\n===== HASIL SVM =====")

    print("Accuracy :", svm_acc)

    print(
        classification_report(
            y_test,
            pred_svm,
            labels=all_labels,
            zero_division=0
        )
    )

    # ========================================================
    # KNN
    # ========================================================

    knn = KNeighborsClassifier(
        n_neighbors=3
    )

    knn.fit(X_train, y_train)

    pred_knn = knn.predict(X_test)

    knn_acc = accuracy_score(
        y_test,
        pred_knn
    )

    print("\n===== HASIL KNN =====")

    print("Accuracy :", knn_acc)

    print(
        classification_report(
            y_test,
            pred_knn,
            labels=all_labels,
            zero_division=0
        )
    )

    # ========================================================
    # CONFUSION MATRIX
    # ========================================================

    cm = confusion_matrix(
        y_test,
        pred_svm,
        labels=all_labels
    )

    disp = ConfusionMatrixDisplay(
        confusion_matrix=cm,
        display_labels=all_labels
    )

    disp.plot(cmap='Blues')

    plt.title("Confusion Matrix")

    plt.show()

    return svm_acc

# ============================================================
# PCA MATCHING
# ============================================================

def pca_matching(des1, des2, n_components):

    if len(des1) == 0 or len(des2) == 0:
        return 0

    combined = np.vstack([des1, des2])

    n_components = min(
        n_components,
        combined.shape[0],
        combined.shape[1]
    )

    pca = PCA(n_components=n_components)

    reduced = pca.fit_transform(combined)

    des1_pca = reduced[:len(des1)]
    des2_pca = reduced[len(des1):]

    bf = cv2.BFMatcher(cv2.NORM_L2)

    matches = bf.knnMatch(
        np.float32(des1_pca),
        np.float32(des2_pca),
        k=2
    )

    good = []

    for pair in matches:

        if len(pair) == 2:

            m, n = pair

            if m.distance < 0.75 * n.distance:
                good.append(m)

    return len(good)

# ============================================================
# MAIN
# ============================================================

if __name__ == "__main__":

    # ========================================================
    # STEP 1
    # ========================================================

    generate_dataset()

    # ========================================================
    # STEP 2
    # ========================================================

    images, labels, paths = load_dataset()

    print("\nJumlah gambar :", len(images))

    # ========================================================
    # STEP 3
    # ========================================================

    methods = [
        "SIFT",
        "ORB",
        "SURF"
    ]

    results = {}

    for method in methods:

        result = extract_features(
            images,
            method
        )

        if result is not None:

            kp_list, des_list = result

            results[method] = {
                "kp": kp_list,
                "des": des_list
            }

    # ========================================================
    # STEP 4
    # VISUALISASI KEYPOINT
    # ========================================================

    sample = images[0]

    for method in results:

        visualize_keypoints(
            sample,
            results[method]["kp"][0],
            f"Keypoints {method}"
        )

    # ========================================================
    # STEP 5
    # FEATURE MATCHING
    # ========================================================

    for method in results:

        print(f"\n===== MATCHING {method} =====")

        kp1 = results[method]["kp"][0]
        kp2 = results[method]["kp"][1]

        des1 = results[method]["des"][0]
        des2 = results[method]["des"][1]

        bf_matches = bf_matching(
            method,
            des1,
            des2
        )

        print("BF Matches :", len(bf_matches))

        flann_matches = flann_matching(
            method,
            des1,
            des2
        )

        print("FLANN Matches :", len(flann_matches))

        H, mask = ransac_homography(
            kp1,
            kp2,
            bf_matches
        )

        if mask is not None:

            print(
                "Jumlah Inlier :",
                np.sum(mask)
            )

        draw_matches(
            images[0],
            kp1,
            images[1],
            kp2,
            bf_matches,
            mask
        )

        precision_recall_analysis(
            bf_matches
        )

    # ========================================================
    # STEP 6
    # BOVW
    # ========================================================

    if "SIFT" in results:

        sift_des = results["SIFT"]["des"]

        vocab_sizes = [10,20,50,100]

        accuracies = []

        for k in vocab_sizes:

            print(f"\n===== BOVW K={k} =====")

            kmeans = build_bovw(
                sift_des,
                k
            )

            X = []

            for des in sift_des:

                hist = extract_histogram(
                    kmeans,
                    des,
                    k
                )

                X.append(hist)

            X = np.array(X)

            acc = classification(
                X,
                labels
            )

            accuracies.append(acc)

        # ====================================================
        # GRAFIK VOCAB SIZE
        # ====================================================

        plt.figure(figsize=(7,5))

        plt.plot(
            vocab_sizes,
            accuracies,
            marker='o'
        )

        plt.xlabel("Vocabulary Size")

        plt.ylabel("Accuracy")

        plt.title("Pengaruh Vocabulary Size")

        plt.grid()

        plt.show()

    # ========================================================
    # STEP 7
    # PCA
    # ========================================================

    if "SIFT" in results:

        des1 = results["SIFT"]["des"][0]
        des2 = results["SIFT"]["des"][1]

        components = [16,32,64,128]

        scores = []

        for c in components:

            score = pca_matching(
                des1,
                des2,
                c
            )

            scores.append(score)

            print(f"\nPCA {c}")

            print(
                "Good Matches :",
                score
            )

        # ====================================================
        # GRAFIK PCA
        # ====================================================

        plt.figure(figsize=(7,5))

        plt.plot(
            components,
            scores,
            marker='o'
        )

        plt.xlabel("PCA Components")

        plt.ylabel("Good Matches")

        plt.title("PCA vs Matching Accuracy")

        plt.grid()

        plt.show()

    # ========================================================
    # STEP 8
    # ANALISIS
    # ========================================================

    analysis = pd.DataFrame({

        "Metode":
        ["SIFT", "ORB", "SURF"],

        "Kecepatan":
        ["Lambat", "Cepat", "Sedang"],

        "Akurasi":
        ["Tinggi", "Sedang", "Tinggi"],

        "Robustness":
        ["Sangat Baik", "Cukup", "Baik"]

    })

    print("\n===== ANALISIS METODE =====")

    print(analysis)

    print("\nProgram selesai.")
```
